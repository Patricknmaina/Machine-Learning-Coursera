- When the discount factor is increased to 0.9, all the states will move to the left towards the bigger reward function. This is because the optimal function in all the states produce a bigger reward towards the left, rather than to the right.

- When the right reward function is reduced to 10, all the states will move to the bigger reward function.

- Setting the misstep probability to 0.1 indicates that there is a 10% chance the rover will move in the wrong direction from the given instruction. It also changes the 5th state from moving towards the higher rewad function to moving towards the lower reward function.

- Increasing the misstep probability will only increase the uncertainity of the movement of the rover by producing different reward values at each state.
